```python
audio = audio.rename(columns={'time_spend': 'audio_time_spend'})
video = video.rename(columns={'time_spend': 'video_time_spend'})
text = text.rename(columns={'time_spend': 'text_time_spend'})
quiz = quiz.rename(columns={'time_spend': 'quiz_time_spend'})
```

```python
audio = audio.set_index(['user_id', 'content_id'])
video = video.set_index(['user_id', 'content_id'])
text = text.set_index(['user_id', 'content_id'])
quiz = quiz.set_index(['user_id', 'content_id'])
```

```python
def to_timedelta(s):
    s = s.split()
    d = int(s[0])
    h, m, s = map(int, s[-1].split(':'))
    
    return pd.Timedelta(weeks=0, days=d, h=h, m=m, sec=s)
```

```python
general = video.join(audio)
general = general.join(text)
general = general.join(quiz)
general['audio_time_spend'] = general.audio_time_spend.fillna(pd.Timedelta(0)).map(lambda x: pd.Timedelta(x))
general['text_time_spend'] = general.text_time_spend.fillna(pd.Timedelta(0)).map(lambda x: pd.Timedelta(x))
general['video_time_spend'] = general.video_time_spend.fillna(pd.Timedelta(0)).map(lambda x: pd.Timedelta(x))
general['quiz_time_spend'] = general.quiz_time_spend.fillna(pd.Timedelta(0)).map(lambda x: pd.Timedelta(x))
```

```python
general['time_spend'] = general['audio_time_spend'] + general['text_time_spend'] + general['video_time_spend'] 
general['time_spend'] = general['time_spend'] + general['quiz_time_spend']
```
