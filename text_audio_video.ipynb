```python
text = pd.read_csv('texttimecount.csv')
video = pd.read_csv('videotimecount.csv')
audio = pd.read_csv('audiotimecount.csv')
```

```python
text.columns, video.columns, audio.columns
```

```python
text = text.rename(columns={'reader_id': 'user_id', 'time_spent':'time_spend'})
audio = audio.rename(columns={'viewer_id': 'user_id', 'viewtime': 'time_spend'})
```

```python
video = video.drop('title', axis=1)
video
```

```python
audio
```

```python
audio = audio.rename(columns={'time_spend': 'audio_time_spend'})
video = video.rename(columns={'time_spend': 'video_time_spend'})
text = text.rename(columns={'time_spend': 'text_time_spend'})
```

```python
audio = audio.set_index(['user_id', 'content_id'])
video = video.set_index(['user_id', 'content_id'])
text = text.set_index(['user_id', 'content_id'])
```

```python
# audio = audio.set_index('user_id')
# video = video.set_index('user_id')
# text = text.set_index('user_id')
```

```python
general = video.join(audio)
general = general.join(text)
general['audio_time_spend'] = general.audio_time_spend.fillna(pd.Timedelta(0))
general['text_time_spend'] = general.text_time_spend.fillna(pd.Timedelta(0))
general['video_time_spend'] = general.video_time_spend.fillna(pd.Timedelta(0))
```

```python
general['time_spend'] = general['audio_time_spend'] + general['text_time_spend'] + general['video_time_spend']
```

```python
general.to_csv('text_audio_video_time_spend.csv')
```
