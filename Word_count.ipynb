```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
basePath = r''
def read(path_to_file):
    return pd.read_csv(f'{base_path}{path_to_file}', encoding='utf16', sep=',')
```

```python
allCoursesInfo = read('dataset/contents.csv')[:1000]
```

```python
allCoursesInfo = allCoursesInfo.dropna()
allCoursesInfo
```

```python
vectorizer = TfidfVectorizer()

names = allCoursesInfo.loc[:, 'normal_name'].values
description = allCoursesInfo.loc[:, 'bdescription'].values

names_weights = vectorizer.fit_transform(names)
description_weights = vectorizer.fit_transform(description)
```

```python
import scipy.sparse

scipy.sparse.save_npz('name_weights.npz', names_weights)
scipy.sparse.save_npz('description_weights.npz', description_weights)
```
