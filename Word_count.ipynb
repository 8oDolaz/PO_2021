```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
basePath = r''
def read(path_to_file):
    return pd.read_csv(f'{base_path}{path_to_file}', encoding='utf16', sep=',')
```

```python
allCoursesInfo = read('dataset/contents.csv')[:10]
```

```python
allCoursesInfo = allCoursesInfo.dropna()
allCoursesInfo
```

```python
names = allCoursesInfo.loc[:, 'normal_name'].values
description = allCoursesInfo.loc[:, 'bdescription'].values
```

```python
import string

def delete_punctuation(array):
    for iteration in range(len(array)):
        for punc in string.punctuation:
            array[iteration] = array[iteration].replace(punc, '')
            
    return array

names = delete_punctuation(names)
description = delete_punctuation(description)
```

```python
import pymorphy2 as pm

morph = pm.MorphAnalyzer()

def back_to_normal(array):
    for element_i, element in enumerate(array):
        element = str(element).split(' ')
        result = []
        for word in element:
            result.append(morph.parse(word)[0].normal_form)
            array[element_i] = str(result)
    
    return array

names = back_to_normal(names)
description = back_to_normal(description)
```

```python
vectorizer = TfidfVectorizer()

names_weights = vectorizer.fit_transform(names)
description_weights = vectorizer.fit_transform(description)
description_weights.shape
```
